```python
import matplotlib.pyplot as plt
import random

# Cases 1-4 are OK
# random.triangular(low, high, mode)
# random.triangular(high, low, mode)
# random.triangular(low, high)
# random.triangular(high, low)

# Case 5 and 6 malfunction, with no warnings/errors
# random.triangular(low, mode, high)
# random.triangular(high, mode, low)
```

```python
# CASE 1 - OK
# random.triangular(low, high, mode)

my_randoms = []
for i in range (100000):
    my_randoms.append(random.triangular(1, 3, 2))
h = plt.hist(my_randoms, bins=200, normed=True)
plt.show()
```

```python
# CASE 2 - OK
# random.triangular(high, low, mode)

my_randoms = []
for i in range (100000):
    my_randoms.append(random.triangular(3, 1, 2))
h = plt.hist(my_randoms, bins=200, normed=True)
plt.show()
```

```python
# CASE 3 - OK
# random.triangular(low, high)

my_randoms = []
for i in range (100000):
    my_randoms.append(random.triangular(1, 3))
h = plt.hist(my_randoms, bins=200, normed=True)
plt.show()
```

```python
# CASE 4 - OK
# random.triangular(high, low)

my_randoms = []
for i in range (100000):
    my_randoms.append(random.triangular(3, 1))
h = plt.hist(my_randoms, bins=200, normed=True)
plt.show()
```

```python
# CASE 5 - NOK (upward slope from 1 to ~2.4)
# random.triangular(low, mode, high)
# (This is numpy.random.triangular input format)

my_randoms = []
for i in range (100000):
    my_randoms.append(random.triangular(1, 2, 3))
h = plt.hist(my_randoms, bins=200, normed=True)
plt.show()
```

```python
# CASE 6 - NOK (downward slope from ~1.6 to 3)
# random.triangular(high, mode, low)

my_randoms = []
for i in range (100000):
    my_randoms.append(random.triangular(3, 2, 1))
h = plt.hist(my_randoms, bins=200, normed=True)
plt.show()
```
